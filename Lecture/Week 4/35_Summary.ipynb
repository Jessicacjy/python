```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load DataFrame
names = pd.read_pickle(r'.\DATA\names\yob_all.pkl')
names.head()
```

### Group By
The process involving one or more of the following steps:
* Splitting the data into groups based on some criteria
* Applying a function to each group independently
* Combining the results into a data structure

```python
#GROUP BY
names.groupby('year').count()
```

```python
names.groupby(['year','sex']).sum()  #count()
```

```python
#APPLY functions
names[['year','births']].apply(lambda x: x.max() - x.min())
```

```python
names.groupby(['year'])['births'].apply(lambda x: x.max() - x.min()) 
```

```python
names.groupby(['year','sex']).apply(lambda x: x.max())
```

```python
names.groupby(['year','sex'])['births'].sum()
```

### Pivot Tables

```python
birthrate = names.pivot_table('births',index='year',columns='sex',aggfunc=sum)
birthrate
```

```python
birthrate.plot(title='Total birth by year and sex')
```
