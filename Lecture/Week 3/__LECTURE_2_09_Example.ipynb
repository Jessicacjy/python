```python
import requests

url = 'http://www.showmeboone.com/sheriff/JailResidents/JailResidents.asp'
response = requests.get(url)
html = response.content.decode('ascii')
print(html)
type(html)
```

```python
import re

#patt = '<td class="one td_left" nowrap="nowrap"></td>'
patt = '<td class="two td_left" nowr>ap="nowrap"(\w*)</td>'
myList = re.findall(patt,html)
myList
```

```python
import re
patt = '<td class="two td_left" nowrap="nowrap">(\w*)</td>' + '\n\s*' + \
    '<td class="two td_left" nowrap="nowrap">(\w*)</td>' + '\n\s*' + \
    '<td class="two td_left" nowrap="nowrap">(\w*)</td>' + '\n\s*' + \
    '<td class="two td_left" nowrap="nowrap">(\w*)</td>' + '\n\s*' + \
    '<td class="two td_left" nowrap="nowrap">(\w*)</td>' # + '\n\s*' + \
    #'<td class="two td_left" nowrap="nowrap">(.*)</td>' + '\n\s*' + \
    #'<td class="two td_right" nowrap="nowrap">(\w*)</td>' + '\n\s*' + \
    #'<td class="two td_left" nowrap="nowrap">(\w*)</td>' 
myList = re.findall(patt,html)
myList
```

```python
[x[3] for x in myList].count('F')
```

```python
[x[3] for x in myList]
```

```python
import numpy as np
import matplotlib.pyplot as plt

%pylab inline
alphab = ['Male', 'Female']
frequencies = [[x[3] for x in myList].count('M'),[x[3] for x in myList].count('F')]

pos = np.arange(len(alphab))
width = 1.0     # gives histogram aspect to the bar diagram

ax = plt.axes()
ax.set_xticks(pos + (width / 2))
ax.set_xticklabels(alphab)

plt.bar(pos, frequencies, width, color='r')
plt.show()
plt.savefig('example02.png')
```

```python
# Import smtplib for the actual sending function
from smtplib import SMTP_SSL as SMTP 

# Here are the email package modules we'll need
from email.mime.image import MIMEImage
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

import getpass

# Create the container (outer) email message.
msg = MIMEMultipart()
msg['Subject'] = 'Stat Report'
msg['From'] = 'anany@sharpedgetech.com'
msg['To'] = 'anan.yaagoub@northwestern.edu'
#msg.preamble = 'Our family reunion'

fp =  open('example02.png', 'rb') 
img = MIMEImage(fp.read())
msg.attach(img)

# Send the email via our own SMTP server.
s = SMTP('smtpout.secureserver.net')
password = getpass.getpass() 
s.login('anany@sharpedgetech.com',password)
s.send_message(msg)
s.quit()
```
